```python
from sklearn.cluster import KMeans
import numpy as np
from skimage import io
from skimage.transform import resize
import os
import matplotlib.pyplot as plt

train_path = "C:/Users/vce/.spyder-py3/archive (5)/train/"
test_path = "C:/Users/vce/.spyder-py3/archive (5)/test/"

train_benign_path = os.path.join(train_path, 'benign')
train_malignant_path = os.path.join(train_path, 'malignant')
test_benign_path = os.path.join(test_path, 'benign')
test_malignant_path = os.path.join(test_path, 'malignant')

train_images_benign = [os.path.join(train_benign_path, f) for f in os.listdir(train_benign_path) if f.endswith('.jpg')]
train_images_malignant = [os.path.join(train_malignant_path, f) for f in os.listdir(train_malignant_path) if f.endswith('.jpg')]

num_images_to_load = 10
train_images = train_images_benign[:num_images_to_load] + train_images_malignant[:num_images_to_load]

def load_and_preprocess_image(image_path, target_size=(150, 150)):
    image = io.imread(image_path)
    image_resized = resize(image, target_size)
    return image_resized

images = []
for img_path in train_images:
    try:
        image = load_and_preprocess_image(img_path)
        images.append(image)
    except Exception as e:
        print(f"Error loading image {img_path}: {e}")

data = np.array([img.reshape(-1, 3) for img in images])
data = np.concatenate(data, axis=0)

if data.shape[0] > 0:
    num_clusters = 3
    kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)  # n_init parametresini açıkça belirtiyoruz
    kmeans.fit(data)
    labels = kmeans.labels_
    print("K-means clustering successful.")
else:
    print("Error: No data to fit K-means.")
```

```python
clustered_images = kmeans.cluster_centers_[labels].reshape(len(images), 150, 150, 3)

import matplotlib.pyplot as plt

num_images_to_visualize = 10

fig, axes = plt.subplots(1, num_images_to_visualize, figsize=(15, 6))
axes = axes.ravel()

for i in range(num_images_to_visualize):
    axes[i].imshow(clustered_images[i])
    axes[i].axis('off')

plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import os

# Eğitim ve doğrulama dizinlerini belirleyelim
train_dir = "C:/Users/vce/.spyder-py3/archive (5)/train/"
validation_dir = "C:/Users/vce/.spyder-py3/archive (5)/test/"

# ImageDataGenerator ile veri artırma işlemleri
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Piksel değerlerini [0,1] aralığına çeker
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

validation_datagen = ImageDataGenerator(rescale=1./255)

# Eğitim verisini yükleme
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='binary')

# Doğrulama verisini yükleme
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='binary')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=10,
    validation_data=validation_generator,
    validation_steps=len(validation_generator))
```

```python
import matplotlib.pyplot as plt

# Eğitim sürecini görselleştirme
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# Başarımları çizme
plt.plot(epochs, acc, 'bo', label='Eğitim Başarımı')
plt.plot(epochs, val_acc, 'b', label='Doğrulama Başarımı')
plt.title('Eğitim ve Doğrulama Başarımı')
plt.xlabel('Epochs')
plt.ylabel('Başarımlar')
plt.legend()

plt.figure()

# Kayıpları çizme
plt.plot(epochs, loss, 'bo', label='Eğitim Kaybı')
plt.plot(epochs, val_loss, 'b', label='Doğrulama Kaybı')
plt.title('Eğitim ve Doğrulama Kaybı')
plt.xlabel('Epochs')
plt.ylabel('Kayıplar')
plt.legend()

plt.show()
```
